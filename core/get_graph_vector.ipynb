```python
# torch_geometric
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import TransformerConv, TopKPooling
from torch_geometric.nn import global_mean_pool as gap, global_max_pool as gmp

class GraphEncode(nn.Module):
    def __init__(self, in_feats, edge_feats, out_feats):
        super(GraphEncode, self).__init__()
        
        self.embedding_size = 64  # 嵌入层维度
        self.n_heads = 4  # 注意力头数
        self.dropout_rate = 0.5
        self.top_k_ratio = 0.5
        self.top_k_every_n = 1
        self.dense_neurons = 128
        self.edge_dim = edge_feats
        
        # 初始化 TransformerConv 层
        self.conv1 = TransformerConv(
            in_feats, 
            self.embedding_size, 
            heads=self.n_heads, 
            dropout=self.dropout_rate,
            edge_dim=self.edge_dim,
            beta=True
        )
        self.transf1 = nn.Linear(self.embedding_size * self.n_heads, self.embedding_size)
        self.bn1 = nn.BatchNorm1d(self.embedding_size)

        # 定义后续的 TransformerConv 层
        self.conv_layers = nn.ModuleList()
        self.transf_layers = nn.ModuleList()
        self.bn_layers = nn.ModuleList()
        self.pooling_layers = nn.ModuleList()

        for _ in range(3):  # 使用 3 层
            self.conv_layers.append(TransformerConv(
                self.embedding_size, 
                self.embedding_size, 
                heads=self.n_heads, 
                dropout=self.dropout_rate,
                edge_dim=self.edge_dim,
                beta=True
            ))
            self.transf_layers.append(nn.Linear(self.embedding_size * self.n_heads, self.embedding_size))
            self.bn_layers.append(nn.BatchNorm1d(self.embedding_size))
            self.pooling_layers.append(TopKPooling(self.embedding_size, ratio=self.top_k_ratio))

    def forward(self, x, edge_attr, edge_index, batch_index):
        print(f"x shape: {x.shape}")
        print(f"edge_attr shape: {edge_attr.shape}")
        # 初始卷积层
        x = self.conv1(x, edge_index, edge_attr)
        x = torch.relu(self.transf1(x))
        x = self.bn1(x)

        global_representation = []
        
        for i, conv_layer in enumerate(self.conv_layers):
            x = conv_layer(x, edge_index, edge_attr)
            x = torch.relu(self.transf_layers[i](x))
            x = self.bn_layers[i](x)
            
            # 聚合图表示
            if i % self.top_k_every_n == 0 or i == len(self.conv_layers) - 1:
                x, edge_index, edge_attr, batch_index, _, _ = self.pooling_layers[i](
                    x, edge_index, edge_attr, batch_index
                )
                global_representation.append(torch.cat([
                    gmp(x, batch_index), 
                    gap(x, batch_index)
                ], dim=1))

        x = sum(global_representation)
        
        return x

class TimePredictionModel(nn.Module):
    def __init__(self, in_feats, edge_feats, out_feats):
        super(TimePredictionModel, self).__init__()
        self.graph_encoder = GraphEncode(in_feats, edge_feats, out_feats)
        self.fc = nn.Linear(2 * 64, 1)  # 2 * 64 是因为我们在全局池化时合并了 mean 和 max 结果
    
    def forward(self, x, edge_attr, edge_index, batch_index):
        graph_embedding = self.graph_encoder(x, edge_attr, edge_index, batch_index)
        return self.fc(graph_embedding)
```

```python
import torch
from torch_geometric.data import Data
import dgl

def dgl_to_pyg(dgl_graph):
    # 从 DGL 图获取信息
    x = dgl_graph.ndata['feature']  # 节点特征
    edge_index = dgl_graph.edges()  # 边索引
    edge_attr = dgl_graph.edata['feature']  # 边特征
    batch = torch.zeros(dgl_graph.number_of_nodes(), dtype=torch.long)  # 默认所有节点在同一图中

    # 转换边索引为 PyG 格式
    edge_index = torch.stack(edge_index).long()

    # 创建 PyG Data 对象
    pyg_data = Data(x=x, edge_index=edge_index, edge_attr=edge_attr, batch=batch)
    return pyg_data

sql_graphs_pyg = []

database.setup(dbname='indexselection_tpcds___1', user='postgres', password='password', host='127.0.0.1', port='5432', cache=False)
cnt = 1

for sql_query in df['query'].values[5000:5050]:
    # print(sql_query)
    if cnt % 10 == 0:
        print(cnt)

    sql_instance = Sql(sql_query)
    # 构建异构图
    g, data_dict, node_indexes = sql_instance.to_hetero_graph_dgl()
    
    # combined_features
    # node
    filter_features = g.ndata['filter']
    edge_features = g.ndata['edge']
    onehot_features = g.ndata['onehot']
    others_features = g.ndata['others'].view(g.ndata['filter'].shape[0], -1)

    # 假设两个特征的维度是相同的
    combined_features = torch.cat((filter_features, edge_features, onehot_features, others_features), dim=1)
    g.ndata['feature'] = combined_features

    # edge
    num_edge_features = 4
    edge_features = torch.randn(g.num_edges(), num_edge_features)
    g.edata['feature'] = edge_features

    # 转换为 PyG 数据对象
    pyg_data = dgl_to_pyg(g)
    sql_graphs_pyg.append(pyg_data)
    
    cnt += 1
```

```python
from torch_geometric.loader import DataLoader

# 创建 PyG DataLoader
data_loader = DataLoader(sql_graphs_pyg, batch_size=32, shuffle=True)

# 初始化模型和优化器
model_params = {
    "model_embedding_size": 64,
    "model_attention_heads": 4,
    "model_layers": 3,
    "model_dropout_rate": 0.5,
    "model_top_k_ratio": 0.5,
    "model_top_k_every_n": 1,
    "model_dense_neurons": 128,
    "model_edge_dim": num_edge_features
}

num_node_features=535
model = TimePredictionModel(num_node_features, num_edge_features, 64)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.MSELoss()

# 训练循环
for epoch in range(10):
    model.train()
    total_loss = 0
    for batch in data_loader:
        # 前向传播
        pred_time = model(batch.x, batch.edge_attr, batch.edge_index, batch.batch)
        
        # 假设目标是生成随机数据，你需要用实际的目标数据替换
        target = torch.randn(pred_time.size(0), 1, dtype=torch.float)
        
        # 计算损失
        loss = criterion(pred_time, target)

        # 反向传播和优化
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f'Epoch {epoch + 1}, Loss: {total_loss:.4f}')
```
